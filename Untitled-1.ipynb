# Introduction
Customer churn is a critical challenge for businesses in the telecommunication industry. Churn occurs when customers discontinue their service, leading to direct revenue loss and increased costs associated with acquiring new customers. Understanding the factors that contribute to customer churn enables organizations to take proactive measures to improve customer retention and long-term profitability .
In this project, we are going to use a supervised machine learning approach to predict customer churn based on customer demographics, service, service usage patterns and account information.

# Business Understanding
The primary business objective of this project is to reduce customer churn by accurately identifying customers who are at risk of leaving the service. By predicting churn in advance, the business can implement targeted retention strategies such as personalized offers, service improvements, or proactive customer support.
From a business perspective, failing to identify a customer who is likely to churn is more costly than incorrectly flagging a customer who would have stayed. As a results, the focus of this project is on maximizing recall for the churn class, ensuring that as many at-risk customers as possible are identified.
The insights gained from this model can help decision-makers allocate resources more effectively, improve customer satisfaction and ultimately increase customer lifetime value.

```python
travel_df['churn'] = travel_df['churn'].map({True: 1, False: 0})
```

# Separate Features and Target

```python
x = travel_df.drop('churn', axis=1)
y = travel_df['churn']
```

```python
x = pd.get_dummies(x, drop_first=True)
```

#  Dataset check

```python
x.shape
y.shape
```

# Building the Train-Split + Baseline Model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train, y_train)

y_pred_lr = log_reg.predict(x_test)
print(classification_report(y_test, y_pred_lr))
```

# Train a basic Decision Tree (baseline)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

dt= DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)
print(classification_report(y_test, y_pred_dt))
```

# EXPLANATION
 The Decision Tree model significantly improved recall for churned customers compared to logistic regression. This indicates the model is better at identifying customers who are likely to leave, which is critical for customers who are likely to leave, which is critical for customers retention strategies. While decision trees may be more prone to overfitting, this model provides stronger predictive value for business objective.

# Train a tuned Decision Tree

```python
dt_tuned = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42
)
dt_tuned.fit(x_train, y_train)

y_pred_dt_tuned= dt_tuned.predict(x_test)
print(classification_report(y_test, y_pred_dt_tuned))
```

# EXPLANATION
 Hyperparameters were tuned to reduce overfitting by limiting tree depth and enforcing minimum sample requirements for splits and leaf nodes.This helps balance model complexity and generalization perfomance.

```python
import pandas as pd

feature_importance = pd.Series(
    dt_tuned.feature_importances_,
    index= x.columns
).sort_values(ascending=False)
feature_importance.head(10)
```

# Model Evaluation
The perfomance of multiple classification models was evaluated using precision, recall, F1-score and accuracy. Since this is a customer churn prediction problem, special attention was given to the model's ability to correctly identify customers who are likely to churn.

# Evaluation Metrics Used 
Accuracy: Overall correctness of the model.
Precision: How many predicted churn cases were actually churn.
Recall: How many actual churn cases were identified.
F1-score: Balance between precision and recall.

# Model Comparison
# Logistic Regression (Baseline Model)
Achieved reasonable overall accuracy.
However, it showed low recall for churned customers, meaning many customers who actually churned were not identified by the model.
This limits its usefulness in a real business context, where missing at-risk customers can lead to revenue loss.

# Decision Tree (Baseline)
Improved recall for churn compared to logistic Regression.
Better captured non-linear relationships in customer behavior.
Risk of overfitting when left unconstrained.

# Tuned Decision Tree (Final Model)
Provided the best balance between recall and overall perfomance.
Hyperparameter tuning reduced overfitting while maintaining strong predictive power.
Demonstrated improved identification of churned customers compared to the baseline models.

# Why does Recall is the most important Metric
In the customer churn prediction, recall for the churn class is more important than overall accuracy.
Failing to identify a customer who is likely to churn means losing the opportunity to intervene with retention strategies.
The tuned Decision Tree model prioritizes recall, making it more suitable for this business problem.

Based on the evaluation results, the tuned Decison Tree was selected as the final model.
